```python
import matplotlib.pyplot as pt
x = ["Day1", "Day2", "Day3", "Day4", "Day5", "Day6"]
y1 = [300, 380, 270, 310, 320, 430]
y2 = [430, 380, 310, 340, 420, 530]

pt.plot(x,y1, marker = "h", ls = ":", color = "green", label ="Week1" )
pt.plot(x,y2, marker = "h", ls = "--", color = "red", label ="Week2",  alpha = .50)
pt.xlabel("Days" )
pt.ylabel("People Visited")
pt.title("Weekly Traffic")
pt.legend()
pt.show()
```
